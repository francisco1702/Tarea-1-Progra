#   **2. Desarrollo Programar Caminos 1**

```python
class PCB:
    def __init__(self, n, m):
        self.n = n
        self.m = m

    def iterativo(self):
        matriz_caminos = [[0] * self.m for _ in range(self.n)]
        for i in range(self.n):
            matriz_caminos[i][0] = 1
        for j in range(self.m):
            matriz_caminos[0][j] = 1
        for i in range(1, self.n):
            for j in range(1, self.m):
                matriz_caminos[i][j] = matriz_caminos[i - 1][j] + matriz_caminos[i][j - 1]
        return matriz_caminos[self.n - 1][self.m - 1]

    def recursivo(self, x=0, y=0):
        if x == self.n - 1 and y == self.m - 1:
            return 1
        caminos = 0
        if x < self.n - 1:
            caminos += self.recursivo(x + 1, y)
        if y < self.m - 1:
            caminos += self.recursivo(x, y + 1)
        return caminos

pcb = PCB(4, 8)
print("N煤mero de caminos (Iterativo):", pcb.iterativo())
print("N煤mero de caminos (Recursivo):", pcb.recursivo())
```

#### Se tiene que las dos soluciones propuestas son de la forma de iteraci贸n y otra con una funci贸n de recursi贸n. Para la primera, se crea una matriz inicializada con ceros. Luego, establece valores de la primera fila y la primera columna en 1, ya que solo hay un camino que pueda seguir. Despu茅s, este m茅todo llena las celdas sumando los valores de la celda de arriba y la de la izquierda, ya que solo se puede mover en las direcciones para arriba y a la derecha. Para terminar, se debe ver el valor de la 煤ltima celda, obteniendo la acumulaci贸n de las sumas de los caminos de las celdas anteriores.
#### Para la parte de recursi贸n, b谩sicamente funciona llamando a la misma funci贸n cada vez que se emplea, para as铆 determinar si lleg贸 al destino o debe seguir corriendo el c贸digo. En cada llamada, el m茅todo verifica si ha alcanzado la celda de destino, en cuyo caso retorna 1 indicando que ha encontrado un camino v谩lido. Si no ha llegado a la celda de destino, verifica si puede moverse hacia abajo o hacia la derecha. #   **2. Desarrollo Programar Caminos 2**

```python
import time
import matplotlib.pyplot as plt

tiempos_de_ejecucion = {}

def almacenar_tiempo(func):
    def envoltura(*args, **kwargs):
        inicio = time.time()
        resultado = func(*args, **kwargs)
        fin = time.time()
        tiempo = fin - inicio
        if func.__name__ not in tiempos_de_ejecucion:
            tiempos_de_ejecucion[func.__name__] = []
        tiempos_de_ejecucion[func.__name__].append(tiempo)
        return resultado
    return envoltura

class CaminosPCB:
    def __init__(self, n, m):
        self.n = n  
        self.m = m  

    def iterativo(self):
        matriz_caminos = [[0] * self.m for _ in range(self.n)]
        for i in range(self.n):
            matriz_caminos[i][0] = 1  
        for j in range(self.m):
            matriz_caminos[0][j] = 1  
        
        for i in range(1, self.n):
            for j in range(1, self.m):
                matriz_caminos[i][j] = matriz_caminos[i-1][j] + matriz_caminos[i][j-1]
        
        return matriz_caminos[self.n-1][self.m-1]

    def recursivo(self, i=None, j=None):
        if i is None or j is None:
            i, j = self.n - 1, self.m - 1
        
        if i == 0 or j == 0:
            return 1
        return self.recursivo(i-1, j) + self.recursivo(i, j-1)

    @almacenar_tiempo  
    def calcular_caminos(self, metodo='iterativo'):
        if metodo == 'iterativo':
            return self.iterativo()
        elif metodo == 'recursivo':
            return self.recursivo()
def grafico():
    tama帽os = range(1, 14)  
    tiempos_iterativo = []
    tiempos_recursivo = []

    for n in tama帽os:
        caminos = CaminosPCB(n, n)        
        caminos.calcular_caminos(metodo='iterativo')
        tiempos_iterativo.append(tiempos_de_ejecucion['calcular_caminos'][-1])        
        caminos.calcular_caminos(metodo='recursivo')
        tiempos_recursivo.append(tiempos_de_ejecucion['calcular_caminos'][-1])
    
    plt.figure(figsize=(10, 6))
    plt.plot(tama帽os, tiempos_iterativo, label='Iterativo', marker='o')
    plt.plot(tama帽os, tiempos_recursivo, label='Recursivo', marker='o')
    plt.title('Tiempo de Ejecuci贸n de M茅todos Iterativo y Recursivo')
    plt.xlabel('Tama帽o de la cuadr铆cula')
    plt.ylabel('Tiempo de ejecuci贸n')
    plt.legend()
    plt.grid(True)
    plt.savefig('tiempos_de_ejecucion.svg', format='svg')
    plt.show()

grafico()

caminos = CaminosPCB(9, 10)  

resultado_iterativo = caminos.calcular_caminos(metodo='iterativo')
print(f"N煤mero de caminos (iterativo): {resultado_iterativo}")

resultado_recursivo = caminos.calcular_caminos(metodo='recursivo')
print(f"N煤mero de caminos (recursivo): {resultado_recursivo}")

print(tiempos_de_ejecucion)
```

En cambio, el m茅todo iterativo evita repetir c谩lculos y no llamar a la misma funci贸n, lo que lo hace mucho m谩s r谩pido y eficiente.
